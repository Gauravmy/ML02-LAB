```python
import pandas as pd

# Series
s = pd.Series([10, 20, 30, 40])
print(s)

# DataFrame
df = pd.DataFrame({
    "Name": ["Amit", "Neha", "Raj"],
    "Age": [25, 28, 22],
    "City": ["Delhi", "Mumbai", "Pune"]
})
print(df)
```

```python
print(df.head())      # First 5 rows
print(df.tail())      # Last 5 rows
print(df.info())      # Summary of DataFrame
print(df.describe())  # Stats summary
```

```python
print(df['Name'])         # Select single column
print(df[['Name','City']])# Multiple columns
print(df.iloc[0])         # First row
print(df.iloc[0:2])       # First 2 rows
print(df.loc[0,'Age'])    # Row 0, column Age
```

```python
print(df[df['Age'] > 23])             # Filter
print(df[(df['Age'] > 23) & (df['City']=="Mumbai")])  # Multiple condition
```

```python
import numpy as np
df.loc[1,'Age'] = np.nan

print(df.isnull())       # Check null
print(df.notnull())      # Check not null
print(df.fillna(0))      # Replace NaN with 0
print(df.dropna())       # Drop rows with NaN
```

```python
df2 = df.rename(columns={'Age':'Years'})
print(df2)

df2['City'] = df2['City'].replace({'Delhi':'New Delhi'})
print(df2)
```

```python
print(df.sort_values(by='Age'))      # Sort by column
print(df.sort_index())               # Sort by index
```

```python
df['Age_Square'] = df['Age'].apply(lambda x: x**2)
print(df)

df['Name_Upper'] = df['Name'].map(str.upper)
print(df)
```

```python
print(df.drop('City', axis=1))         # Drop column
print(df.drop_duplicates())            # Remove duplicate rows
```

```python
print(df.dtypes)             # Data types
df['Age'] = df['Age'].astype('float') # Convert type
print(df.dtypes)
```

```python
print(df.groupby("City")['Age'].mean())   # Group by mean
print(df.groupby("City").agg({'Age':['mean','max','min']}))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# STEP 1: Upload Files
print("ðŸ‘‰ Please upload your 3 Excel files...")
uploaded = files.upload()

dfs = []
for fname in uploaded.keys():
    # Try reading as CSV based on previous experience
    try:
        dfs.append(pd.read_csv(fname))
    except Exception as e:
        print(f"Could not read {fname} as CSV: {e}")
        # If CSV reading fails, you might need to inspect the file content
        # or try other engines if you suspect it's a different Excel format.


if len(dfs) < 3:
    print("Could not load all 3 files. Please check the uploaded files.")
else:
    df1, df2, df3 = dfs[0], dfs[1], dfs[2]

    print("\nâœ… All 3 datasets loaded successfully as CSV!\n")

    # STEP 2: Graph Function
    def plot_graph(df, dataset_name):
        plt.figure(figsize=(7,5))
        num_cols = df.select_dtypes(include="number").columns
        cat_cols = df.select_dtypes(include="object").columns

        if len(num_cols) >= 1 and len(cat_cols) >= 1:
            # Barplot (category vs numeric)
            # Using the first numeric and first categorical column as examples
            sns.barplot(x=cat_cols[0], y=num_cols[0], data=df, palette="coolwarm", hue=cat_cols[0], legend=False)
            plt.title(f"{dataset_name} - Barplot of {num_cols[0]} by {cat_cols[0]}")
            plt.xticks(rotation=45)

        elif len(num_cols) >= 1:
            # Histogram of first numeric column
            sns.histplot(df[num_cols[0]], bins=20, kde=True, color="skyblue")
            plt.title(f"{dataset_name} - Histogram of {num_cols[0]}")
            plt.xlabel(num_cols[0])
            plt.ylabel("Frequency")

        elif len(cat_cols) >= 1:
            # Countplot of first categorical column
            sns.countplot(x=cat_cols[0], data=df, palette="Set2", hue=cat_cols[0], legend=False)
            plt.title(f"{dataset_name} - Countplot of {cat_cols[0]}")
            plt.xticks(rotation=45)

        else:
            print(f"âš ï¸ {dataset_name} has no numeric or categorical columns for plotting.")

        plt.show()

    # STEP 3: Plot Graphs for Each Dataset
    plot_graph(df1, "Dataset 1")
    plot_graph(df2, "Dataset 2")
    plot_graph(df3, "Dataset 3")
```
